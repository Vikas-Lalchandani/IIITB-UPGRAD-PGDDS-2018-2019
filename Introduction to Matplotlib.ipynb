# Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
# plot all figures inside the jupyter notebook
%matplotlib inline 
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plots
fig, ax = plt.subplots(figsize = (8,8)) #width, height

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save and Show (you save the whole figure)
fig.savefig("Images/sample-plot.png")
```

### Two options for Subplots

```python
#Subplots Option 1
fig, ((ax0,ax1),(ax2,ax3)) = plt.subplots(nrows = 2,
                                          ncols = 2,
                                          figsize = (20,10))

#Plot to each different axis
ax0.plot(x_new, x_new/2);
ax0.set(xlabel = "x_new",
        ylabel = "x_new/2",
        title = "Linear Relationship")

ax1.scatter(x_new, np.sin(x_new))
ax1.set(xlabel = "x_new",
        ylabel = "Sine of x_new",
        title = "Sine Cuve")

ax2.bar(butter_prices.keys(), butter_prices.values())
ax2.set(title = "Butter Store Prices",
      ylabel = "Butter Price ($)",
      xlabel = "Butter Types")

ax3.hist(np.random.randn(1000))
ax3.set(title = "Histogram",
      ylabel = "Height",
      xlabel = "Bins");
```

```python
#Subplots Option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (20,10))

#Plot to each different index
ax[0,0].plot(x_new, x_new/2);
ax[0,0].set(xlabel = "x_new",
        ylabel = "x_new/2",
        title = "Linear Relationship")

ax[0,1].scatter(x_new, np.sin(x_new))
ax[0,1].set(xlabel = "x_new",
        ylabel = "Sine of x_new",
        title = "Sine Cuve")

ax[1,0].bar(butter_prices.keys(), butter_prices.values())
ax[1,0].set(title = "Butter Store Prices",
      ylabel = "Butter Price ($)",
      xlabel = "Butter Types")

ax[1,1].hist(np.random.randn(1000))
ax[1,1].set(title = "Histogram",
      ylabel = "Height",
      xlabel = "Bins");
```

## Which one you should use? (pyplot Vs matplotlib OO method)

* When plotting something quickly better to pyplot method
* When plotting something advanced, better to use matplotlib OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#pyplot approach
over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target",
             figsize = (10,6));
```

```python
## Object Oriented with pylot
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target",
             ax = ax);

ax.set_xlim([45,100]);
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

#Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

#customize the plot
ax.set(title  = "Heart Disease and Cholestrol levels",
       xlabel = "Age",
       ylabel = "Cholestrol")

#To set the legend
ax.legend(*scatter.legend_elements(), title = "Target")

#to plot the horizontal line on the axis showing mean of the cholestrol values
ax.axhline(over_50["chol"].mean(),
           linestyle = "--",
           lw = 3.0);
```

```python
over_50.head()
```

```python
#subplot of age, chol and thalach

fig,((ax0,ax1)) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10,10),
                               sharex = True); #--> #to share the label for x-axis

#plot data for ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

# Names axes an title for ax0
ax0.set(title = "Heart Disease and Cholestrol levels",
        ylabel = "Cholestrol")

#Legend for ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

#Horizontal Mean line for ax0
ax0.axhline(over_50["chol"].mean(),
            ls = "--",
            lw = 3.0);

#plot data for ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])

# Names axes an title for ax1
ax1.set(title = "Heart Disease and Heart Rates",
        xlabel = "Age",
        ylabel = "Heart Rate")

#Legend for ax0
ax1.legend(*scatter.legend_elements(), title = "Target")

#Horizontal Mean line for ax0
ax1.axhline(over_50["thalach"].mean(),
            ls = "--",
            lw = 3.0)

#title for overall figure
fig.suptitle("Hear Disease Analysis", fontsize = 16, fontweight = "bold");
```

### Customizing the plots

```python
#Setting the style
plt.style.use('seaborn-deep')

#subplot of age, chol and thalach

fig,((ax0,ax1)) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (15,10),
                               sharex = True); #--> #to share the label for x-axis

#plot data for ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"],
                      cmap = "winter")

#set x-axis limit for ax0
ax0.set_xlim([50,80])

# Names axes an title for ax0
ax0.set(title = "Heart Disease and Cholestrol levels",
        ylabel = "Cholestrol")

#Legend for ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

#Horizontal Mean line for ax0
ax0.axhline(over_50["chol"].mean(),
            ls = "--",
            lw = 3.0,
            c = "maroon");

#plot data for ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap = "winter")

# Names axes an title for ax1
ax1.set(title = "Heart Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")

#setting y-axis limit for ax1
ax1.set_ylim([60,200])

#Legend for ax0
ax1.legend(*scatter.legend_elements(), title = "Target")

#Horizontal Mean line for ax0
ax1.axhline(over_50["thalach"].mean(),
            ls = "--",
            lw = 3.0,
            c = "maroon")

#title for overall figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
